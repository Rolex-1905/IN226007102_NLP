## Task 1: Conceptual Understanding

### Q1: Difference between "Love" and "love"
In NLP, "Love" and "love" are treated as different tokens due to case sensitivity. Converting text to lowercase ensures consistency and avoids duplication.

### Q2: What happens if stopwords are not removed?
If stopwords are not removed, the dataset will contain unnecessary words like "is", "the", "and", increasing noise and reducing model efficiency.

### Q3: When removing stopwords can be harmful?
1. Sentiment Analysis: "I am not happy" → removing "not" changes meaning.
2. Question Answering: "What is your name?" → removing "what" loses intent.

### Q4: Stemming vs Lemmatization
Stemming removes suffixes and may not produce real words, while lemmatization returns meaningful base words using vocabulary.

```python
import re
from collections import Counter
```

## Task 2: Preprocessing Function

**Problem Statement:**  
Build a function to clean and preprocess raw text data.

```python
def preprocess_text(text):
    try:
        if not text or not isinstance(text,str):return [],""  # validate input
        text=text.lower()
        text=re.sub(r"http\S+|www\S+","",text)  # remove URLs
        text=re.sub(r"\S+@\S+","",text)  # remove emails
        text=re.sub(r"\d+","",text)  # remove numbers
        text=re.sub(r'(.)\1{2,}',r'\1',text)  # normalize repeated chars
        text=re.sub(r'[^a-z\s]',"",text)  # remove special chars
        text=re.sub(r'\s+'," ",text).strip()  # normalize spaces
        tokens=text.split()
        tokens=[w for w in tokens if len(w)>2 or w in ['no','not']]  # keep meaningful words
        return tokens," ".join(tokens)
    except:return [],""  # fail-safe
```

## Task 3: Stress Testing

Testing the preprocessing function on diverse real-world sentences.

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for sentence in test_sentences:
    tokens, cleaned = preprocess_text(sentence)

    print("Original:", sentence)
    print("Tokens:", tokens)
    print("Cleaned:", cleaned)
    print("-" * 50)

    results.append((sentence, tokens))
```

### Explanation:
The preprocessing function successfully removes noise such as numbers, URLs, emojis, and repeated characters.  
It converts text into clean tokens suitable for further NLP tasks.

## Task 4: Token Analytics

Analyzing token statistics for each sentence.

```python
for sentence,tokens in results:
    if tokens:
        total=len(tokens)
        unique=len(set(tokens))
        avg_len=sum(len(t)for t in tokens)/total  # average token length
    else:total,unique,avg_len=0,0,0  # handle empty case

    print("Sentence:",sentence)
    print("Total Tokens:",total)
    print("Unique Tokens:",unique)
    print("Average Token Length:",round(avg_len,2))
    print("-"*50)
```

## Task 5: Frequency Analysis

Finding most and least frequent words.

```python
all_tokens = []

for _, tokens in results:
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[-5:])
```

## Task 6: Full Pipeline

Combining preprocessing into a reusable pipeline.

```python
def full_pipeline(text_list):
    all_tokens=[]
    clean_sentences=[]
    for text in text_list:
        tokens,cleaned=preprocess_text(text)
        all_tokens.extend(tokens)  # merge all tokens
        clean_sentences.append(cleaned)
        
    return {"tokens":all_tokens,"clean_sentences":clean_sentences}

output=full_pipeline(test_sentences)
print(output)
```

## Task 7: Error Handling

Testing edge cases such as empty strings, emojis, and numbers.

```python
edge_cases = ["", "😂😂😂", "123456"]

for case in edge_cases:
    tokens, cleaned = preprocess_text(case)
    print("Input:", case)
    print("Tokens:", tokens)
    print("Cleaned:", cleaned)
    print("-" * 40)
```

### Explanation:
The function handles edge cases effectively.  
For empty strings, emojis, and numeric inputs, it safely returns empty tokens without errors.
